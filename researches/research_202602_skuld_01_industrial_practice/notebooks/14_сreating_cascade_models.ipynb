```python
from pathlib import Path
from predictors.pipeline import CascadedClassifier
from predictors.serialization import load_model, save_model
```

```python
models_dir = Path("../models")
```

```python
gender_svc_poly = load_model(models_dir / "svc_poly_gender_best" / "model.pkl")
gender_nb = load_model(models_dir / "nb_gender_best" / "model.pkl")

men_age_gbm = load_model(models_dir / "gbm_men_age_best" / "model.pkl")
women_age_gbm = load_model(models_dir / "gbm_women_age_best" / "model.pkl")

men_age_svc_rbf = load_model(models_dir / "svc_rbf_men_age_best" / "model.pkl")
women_age_svc_rbf = load_model(models_dir / "svc_rbf_women_age_best" / "model.pkl")

men_age_lr = load_model(models_dir / "lr_men_age_best" / "model.pkl")
women_age_lr = load_model(models_dir / "lr_women_age_best" / "model.pkl")
```

```python
cascade_1 = CascadedClassifier(gender_svc_poly, [women_age_gbm, men_age_gbm])
cascade_2 = CascadedClassifier(gender_svc_poly, [women_age_svc_rbf, men_age_svc_rbf])
cascade_3 = CascadedClassifier(gender_svc_poly, [women_age_lr, men_age_lr])
cascade_4 = CascadedClassifier(gender_nb, [women_age_gbm, men_age_gbm])
```

```python
for model, name in [
    (cascade_1, "cascade_svc_poly_gbm"), 
    (cascade_2, "cascade_svc_poly_svc_rbf"), 
    (cascade_3, "cascade_svc_poly_lr"), 
    (cascade_4, "cascade_nb_gbm")
]:
    save_model(models_dir, model, name)
```
