# Подготовка исходных данных к анализу

```python
import pandas as pd
```

```python
texts_df = pd.read_parquet("../data/raw/text.parquet")
metrics_df = pd.read_parquet("../data/raw/metrics.parquet")
df = pd.concat([texts_df, metrics_df], axis=1)
```

```python
df.head()
```

```python
start_size = len(df)
```

```python
is_dublicate = df.text.duplicated()
df = df[~is_dublicate]
```

```python
df['length'] = df['text'].apply(len)
```

```python
df.columns
```

```python
df.info()
```

```python
with_na = df.columns[df.isna().sum() != 0].to_list()
display(with_na, len(with_na))
```

## Установка значений по умолчанию

Для метрик значение `NaN` соответствует 0

```python
zero_default_cols = with_na[:]
zero_default_cols.remove('source_url')

df[zero_default_cols] = df[zero_default_cols].fillna(0)
```

## Базовый осмотр статистических показателей

## Удаляем наблюдения
Для обеспечения качества данных мы применяем последовательную фильтрацию, удаляя тексты с явными признаками ошибок разметки или выходящие за границы применимости нашего анализа.

1. Длина текста: от 300 до 10 000 символов (недостаточно информации либо превышение границ применимости)
2. Средняя длина предложения в словоформах: от 3 до 15 (ошибка токенизации либо неинформативный/избыточно сложный текст)
3. Средняя длина предложения в символах: не более 130 (аномально длинные предложения)
4. Среднее количество чисел: не более 2 на предложение
5. Среднее количество перечислений: не более 0.4 на предложение
6. Среднее количество единиц измерения: не более 0.4 на предложение
7. Среднее количество иностранных слов: не более 1 на предложение

```python
df = df[
    (df['length'].between(300, 10000)) &
    (df['text__lexicon_size'].between(3, 15)) &
    (df['text__symbols_length'] <= 130) &
    (df['entities__mean_usage_NUMBER'] <= 2) &
    (df['entities__mean_usage_ENUM'] <= 0.4) &
    (df['entities__mean_usage_MEAS'] <= 0.4) &
    (df['entities__mean_usage_FOREIGN'] <= 1)
]
```

## Исправление наименования признаков

```python
df.rename(columns={
    "pos__('ADJ-Pos', 'NOUN')": "pos__ADJPOS_NOUN",
    "pos__('PUNCT', 'SCONJ')": "pos__PUNCT_SCONJ",
    "pos__('ADP', 'ADJ-Pos')": "pos__ADP_ADJPOS",
    "pos__('ADP', 'PRON')": "pos__ADP_PRON",
    "pos__('PUNCT', 'ADV')": "pos__PUNCT_ADV",
    "pos__('ADJ-Pos', 'PUNCT')": "pos__ADJPOS_PUNCT",
    "pos__('PUNCT', 'CCONJ')": "pos__PUNCT_CCONJ",
    "pos__('PUNCT', 'PRON')": "pos__PUNCT_PRON",
    "pos__('PROPN', 'PUNCT')": "pos__PROPN_PUNCT",
    'punct__mean_usage_.': "punct__mean_usage_dot",
    'punct__mean_usage_,': "punct__mean_usage_comma",
    'punct__mean_usage_:': "punct__mean_usage_colon",
    'punct__mean_usage_"': "punct__mean_usage_quote",
    'punct__mean_usage_?': "punct__mean_usage_question",
    'punct__mean_usage_(': "punct__mean_usage_openparen",
    'punct__mean_usage_)': "punct__mean_usage_closeparen",
    'punct__mean_usage_-': "punct__mean_usage_hyphen",
    'punct__mean_usage_!': "punct__mean_usage_exclamation",
    'punct__mean_usage_&': "punct__mean_usage_ampersand",
    'punct__mean_usage_]': "punct__mean_usage_closebracket",
    'punct__mean_usage_[': "punct__mean_usage_openbracket",
    'punct__mean_usage_;': "punct__mean_usage_semicolon",
    'punct__mean_usage_<': "punct__mean_usage_lessthan",
    'punct__mean_usage_>': "punct__mean_usage_greaterthan",
    'punct__mean_usage_/': "punct__mean_usage_slash",
    'punct__mean_usage_*': "punct__mean_usage_asterisk",
    'punct__mean_usage__': "punct__mean_usage_underscore",
    'punct__mean_usage_~': "punct__mean_usage_tilde"
}, inplace=True)
```
